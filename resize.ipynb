```python
from PIL import Image
import os

def resize_images_in_folder(input_folder, output_folder, new_width, new_height):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Gunakan fallback agar aman di semua versi Pillow
    try:
        resample = Image.Resampling.LANCZOS
    except AttributeError:
        resample = Image.ANTIALIAS  # untuk versi lama

    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        if os.path.isfile(input_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
            try:
                with Image.open(input_path) as img:
                    img_resized = img.resize((new_width, new_height), resample)
                    img_resized.save(output_path)
                    print(f"Resized {filename}")
            except Exception as e:
                print(f"Error resizing {filename}: {e}")

# Konfigurasi folder dan ukuran
input_folder = "/Users/anggun/Downloads/SKRIPSI ANGGUN/ANGGUN/2. Data Baru CLAHE/Data Awal Sebelum Diolah/Pothole"
output_folder = "/Users/anggun/Downloads/SKRIPSI ANGGUN/ANGGUN/2. Data Train & Validation/Pothole"
new_width = 1200
new_height = 1500

resize_images_in_folder(input_folder, output_folder, new_width, new_height)
```
